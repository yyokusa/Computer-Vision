```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
import pyautogui
import time
import os
```

```python
# Shame 18 - 0
# Click Shape and open the game when you hear the "BEEP" sound
# Screen Resolution: 1024 x 640
template1 = cv2.imread('ucgen.png',0)
template2 = cv2.imread('kare.png',0)
template3 = cv2.imread('yildiz.png',0)
template4 = cv2.imread('altigen.png',0)
def detect_key():
    print('\n Called here \n')
    img_rgb = cv2.imread('image.png')
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
    res1 = cv2.matchTemplate(img_gray,template1,cv2.TM_CCOEFF_NORMED)
    res2 = cv2.matchTemplate(img_gray,template2,cv2.TM_CCOEFF_NORMED)
    res3 = cv2.matchTemplate(img_gray,template3,cv2.TM_CCOEFF_NORMED)
    res4 = cv2.matchTemplate(img_gray,template4,cv2.TM_CCOEFF_NORMED)
    threshold = 0.8
    loc1 = np.where( res1 >= threshold)
    loc2 = np.where( res2 >= threshold)
    loc3 = np.where( res3 >= threshold)
    loc4 = np.where( res4 >= threshold)
    if (len(loc3[0]) > 20):
        return 'yildiz'
    elif (len(loc2[0]) > 20):
        return 'kare'
    elif (len(loc4[0]) > 20):
        return 'altigen'
    elif (len(loc1[0]) > 20):
        return 'ucgen'
    else:
        return None
```

```python
time.sleep(2)
os.system( "say beep" )
time.sleep(1)
while True:
    myScreenshot = pyautogui.screenshot(region=(800, 800, 1000, 640))
    myScreenshot.save('image.png')
    result = detect_key()
    print('result: ', result)
    if result == 'ucgen':
        pyautogui.keyDown('shift')
        pyautogui.keyDown('a')
        os.system( "say triangle" )
        time.sleep(0.6)
        pyautogui.keyUp('shift')
        pyautogui.keyUp('a')
        pass
    elif result == 'kare':
        pyautogui.keyDown('shift')
        pyautogui.keyDown('s')
        os.system( "say square" )
        time.sleep(0.7)
        pyautogui.keyUp('shift')
        pyautogui.keyUp('s')
        pass
    elif result == 'yildiz':
        pyautogui.keyDown('shift')
        pyautogui.keyDown('d')
        os.system( "say star" )
        time.sleep(0.6)
        pyautogui.keyUp('shift')
        pyautogui.keyUp('d')
        pass
    elif result == 'altigen':
        pyautogui.keyDown('shift')
        pyautogui.keyDown('f')
        os.system( "say six" )
        time.sleep(0.8)
        pyautogui.keyUp('shift')
        pyautogui.keyUp('f')
        pass
    else:
        pyautogui.keyDown('shift')
        pyautogui.keyDown('l')
        os.system( "say None")
        time.sleep(0.5)
        pyautogui.keyUp('shift')
        pyautogui.keyUp('l')
```
